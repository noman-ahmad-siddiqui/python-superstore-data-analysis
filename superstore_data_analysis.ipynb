## 1 Segment Analysis

### Which segment buys the most?

```python
segment_counts = df.groupby('Segment')['Row ID'].count()
segment_counts
```

```python
segment_counts.index
```

```python
segment_counts.values
```

```python
plt.barh(segment_counts.index, segment_counts.values, height = 0.5, color = 'green')

plt.title("Which segment buys the most?")
plt.xlabel("Purchases")
plt.ylabel("Segment")

# plt.savefig("C:/Users/noman/Downloads/monthly_sales_trend.png", bbox_inches="tight") # To save image
plt.show()
```

### Which segment is most profitable?

```python
df.groupby('Segment')['Profit'].sum()
```

```python
uniqueSegments = df['Segment'].unique()
profitBySegment = df.groupby('Segment')['Profit'].sum()

plt.title("Which segment is most profitable?")
plt.xlabel("Segment")
plt.ylabel("Profit")

plt.bar(uniqueSegments,profitBySegment, color = 'red', width = 0.5)

plt.grid(axis = 'y')

plt.show
```

## 2 Sales Analysis

### Total Sales

```python
TotalSales = df['Sales'].sum()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(4,2))
plt.text(0.5, 0.5, f"Total Sales: {TotalSales}" , fontsize=20, ha='center', color= 'green', fontweight='bold')  # plt.text(x, y, text) e.g x,y are horizontal and vertical positions, ranges from 0 to 1. (0,0) = bottom-left, (1,1) = top-right, (0.5,0.5) = center 
plt.axis('off') # Hides X and Y axis completely, Otherwise, you would see numbers and gridlines
plt.show()
```

```python
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Day of Week'] = df['Order Date'].dt.dayofweek
```

### What is the monthly sales trend?

```python
monthlySales = df.groupby('Order Month')['Sales'].sum()
monthlySales
```

```python
plt.plot(monthlySales.index, monthlySales.values, marker ='D')

plt.savefig("C:/Users/noman/Downloads/monthly_sales_trend.png", bbox_inches="tight")
plt.show()
```

### What is the yearly sales growth? 

```python
yearlyGrowth = df.groupby('Order Year')['Sales'].sum()
yearlyGrowth
```

```python
yearlyGrowth.index
```

```python
plt.plot(yearlyGrowth.index, yearlyGrowth.values,marker='o')
plt.xticks(yearlyGrowth.index) #to get exact years at x-axis

# plt.savefig("C:/Users/noman/Downloads/yearly_sales_growth.png", bbox_inches="tight")
plt.show()
```

### What are the top 10 products by sales?

```python
top10 = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top10
```

```python

plt.figure(figsize=(10,6), facecolor='pink') # 10 = width of the figure in inches, 6 = height of the figure in inches
top10.sort_values().plot(kind='barh')

plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.ylabel("Product Name")

# plt.savefig("C:/Users/noman/Downloads/top_10_products_by_sales.png", bbox_inches="tight")
plt.show()
```

## 3. Profit Analysis

### What is the total profit? 

```python
totalProfit = int(df['Profit'].sum())
totalProfit
```

```python
plt.figure(figsize=(2,2))
plt.text(0.5,0.5, f"Total Profit: {totalProfit}", fontsize=20, color= 'green')
plt.axis('off')
plt.show
```

### How is profit divided across categories?

```python
profitByCategory = df.groupby('Category')['Profit'].sum()
profitByCategory
```

```python
mycolors = ["hotpink", "b", "#4CAF50"] # defining colors

myExplode = [0.1, 0, 0] # defining the first category i.e. Furniture should explode outside with one point

plt.pie(profitByCategory, labels = profitByCategory.index, colors = mycolors, explode =myExplode)
plt.title("Profit By Category")
# plt.legend(title = "Profit Division") # show legends
# plt.savefig("C:/Users/noman/Downloads/profit_by_category.png", bbox_inches="tight")
plt.show()
```

### Which states generate losses?

```python
profitByState = df.groupby('State')['Profit'].sum().sort_values()
lossStates = profitByState[profitByState < 0].sort_values()
lossStates
```

### Which sub-categories are profitable or loss-making? 

```python
plt.figure(figsize=(12,5))
sns.barplot(data=df, x='Sub-Category', y='Profit', estimator=sum)
plt.xticks(rotation=45)
# plt.savefig("C:/Users/noman/Downloads/profit_loss_by_sub_category.png", bbox_inches="tight")
plt.show()
```

## 6. Shipping Analysis

```python
df['Shipping_Days'] = df['Ship Date'] - df['Order Date']
df['Shipping_Days'] = df['Shipping_Days'].dt.days
df['Shipping_Days']
```

```python
df['Shipping_Days'].describe()
```

### How many days does shipping take on average? 

```python
df.groupby('Ship Mode')['Shipping_Days'].mean().sort_values()
```

### Which ship mode is fastest?

```python
df.groupby('Ship Mode')['Shipping_Days'].mean().sort_values().head(1)
```

### Does faster shipping increase sales?

```python
df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending = False)
```
